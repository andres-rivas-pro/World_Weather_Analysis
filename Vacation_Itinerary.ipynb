```python
import gmaps
from config import g_key
import pandas as pd
import requests
```

```python
#Import the WeatherPy_vacation.csv file as a new DataFrame
challenge_itinerary_df = pd.read_csv('data/weatherPy_vacation.csv')
challenge_itinerary_df.head()
```

```python
Brazil_itinerary_df = challenge_itinerary_df.loc[(challenge_itinerary_df['Country'] =='BR')]
Brazil_itinerary_df
```

```python
# choose four cities from Brazil.(jardim, breves, diamantino, anori)
four_cities_df = Brazil_itinerary_df.loc[[21,45,90,107],:]
four_cities_df
```

```python
# create a list of tuple of 4 cities' coordinate.
gmaps.configure(api_key = g_key)

BR_coord_list = list(zip(four_cities_df['Lat'], four_cities_df['Lng']))
BR_coord_list
```

```python
import gmaps.datasets
# Latitude-longitude pairs
jardim = (-21.48, -56.14)
breves = (-1.68, -50.48)
diamanitino = (-14.41, -56.45)

fig = gmaps.figure()
jardim2diamanitino = gmaps.directions_layer(jardim, diamanitino)
fig.add_layer(jardim2diamanitino)
fig
```

```python
gmaps.configure(api_key = g_key)
locations = four_cities_df[['Lat','Lng']]
info_box_template='''
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather</dt><dd>{Current Description} and {Max Temp}°F</dd>
</dl>'''

hotel_info = [info_box_template.format(**row) for index, row in four_cities_df.iterrows() ]
fig = gmaps.figure()
marker_layer = gmaps.marker_layer(locations, info_box_content = hotel_info)
fig.add_layer(marker_layer)

fig
```
